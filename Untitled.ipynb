### Removing images without annotations

```python
def check_bounding_box(img_name):
    img_path=os.path.join('data','images',img_name)
    dest_path=os.path.join("data","cleaned_images",img_name)
    if not os.path.isfile(img_path):
        index=annotations[annotations["image_name"]==img_name].index
        annotations.drop(index,axis=0,inplace=True)
    else:
        shutil.copyfile(img_path,dest_path)
```

```python
for img in annotations["image_name"]:
    check_bounding_box(img)
```

```python
fig2,ax2=plt.subplots(1,5,figsize=(18,10))
for i in range(5):
    img=cv2.imread(os.path.join("data","cleaned_images",multi_bounds[i]))
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    
    #getting bounding boxes for images
    bboxes=annotations[annotations["image_name"]==multi_bounds[i]]
    for _,row in bboxes.iterrows():
        cv2.rectangle(img,(row['x0'],row['y0']),(row['x1'],row['y1']),(255,0,0),3)
    ax2[i].imshow(img)
plt.show()
```

```python
def resize_img_bound(row,target_size=(224,224)):
    img_path=os.path.join("data","cleaned_images",row[0])
    img=cv2.imread(img_path)
    
    # resizing the image
    orig_h,orig_w=img.shape[:2]
    new_h,new_w=target_size
    resized_image=cv2.resize(img,(new_w,new_h))

    # resizing the bounding box
    x_scale=new_w/orig_w
    y_scale=new_h/orig_h
    x0=int(row["x0"]*x_scale)
    y0=int(row["y0"]*y_scale)
    x1=int(row["x1"]*x_scale)
    y1=int(row["y1"]*y_scale)

    # saving resized images
    save_path=os.path.join('data',"resized",row[0])
    os.makedirs("data/resized",exist_ok=True)
    cv2.imwrite(save_path,resized_image)

    return row[0],new_w,new_h,x0,y0,x1,y1
```

```python
resized_data=annotations.apply(resize_img_bound,axis=1)
resized_data.head()
```

```python
resized_df=pd.DataFrame(resized_data.to_list(),columns=annotations.columns)
resized_df.head()
```

## Labeling and structuring

```python
# creating folders for images and labels   
os.makedirs('data/train/images',exist_ok=True)
os.makedirs('data/train/labels',exist_ok=True)
```

```python
for img_name in annotations["image_name"].unique():
    bboxes=annotations[annotations["image_name"]==img_name]

    # moving images to training folder
    src=os.path.join('data',"resized",img_name)
    dst=os.path.join("data","train","images",img_name)
    shutil.copy(src,dst)

    height=bboxes["height"].values[0]
    width=bboxes["width"].values[0]

    label_path=os.path.join("data/train/labels",img_name.replace(".jpg",".txt"))
    with open(label_path, "w") as f:
        for _, row in bboxes.iterrows():
            x_center = ((row["x0"] + row["x1"]) / 2) / width
            y_center = ((row["y0"] + row["y1"]) / 2) / height
            w = (row["x1"] - row["x0"]) / width
            h = (row["y1"] - row["y0"]) / height
            f.write(f"0 {x_center} {y_center} {w} {h}\n")
```

```python

from IPython.display import display, clear_output

# Start webcam
cap = cv2.VideoCapture(0)

# Check if webcam opened correctly
if not cap.isOpened():
    print("Error: Cannot open webcam")
else:
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Failed to grab frame")
                break

            results = model.predict(source=frame, save=False, conf=0.3, verbose=False)

            annotated_frame = results[0].plot()

            # Convert to RGB for matplotlib
            frame_rgb = cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB)

            # Display in notebook
            clear_output(wait=True)
            plt.imshow(frame_rgb)
            plt.axis("off")
            plt.show()

    except KeyboardInterrupt:
        clear_output(wait=True)
        print("Stopped manually.")

    finally:
        cap.release()
```
